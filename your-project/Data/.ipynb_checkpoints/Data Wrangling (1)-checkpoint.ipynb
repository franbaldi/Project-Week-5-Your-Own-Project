```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from requests import get
import re
```

```python
# Scraping web FBREF.com
url = 'https://widgets.sports-reference.com/wg.fcgi?css=1&site=fb&url=%2Fen%2Fcomps%2F22%2FMajor-League-Soccer-Stats&div=div_stats_standard_squads'
# "https://widgets.sports-reference.com/wg.fcgi?css=1&site=fb&url=%2Fen%2Fcomps%2F22%2FMajor-League-Soccer-Stats&div=div_stats_keeper_squads"
# "https://widgets.sports-reference.com/wg.fcgi?css=1&site=fb&url=%2Fen%2Fcomps%2F22%2FMajor-League-Soccer-Stats&div=div_stats_keeper_adv_squads"
# "https://widgets.sports-reference.com/wg.fcgi?css=1&site=fb&url=%2Fen%2Fcomps%2F22%2FMajor-League-Soccer-Stats&div=div_stats_shooting_squads"

html = requests.get(url).content
html_soup = BeautifulSoup(html, 'html.parser')
```

```python
# Soup scraping 
MLS_2020_General_Stats = html_soup.find_all('table', {'class': "min_width sortable stats_table"})
```

```python
head = MLS_2020_General_Stats[0].find('thead')
```

```python
body = MLS_2020_General_Stats[0].find('tbody')
```

```python
list_of_dictionaries = []

for i in body:
    row_list = []
    country = i.find('th').getText()
    row = i.find_all('td')
    row_list.append(country)
    for cell in row:
        row_list.append(cell.getText())
    list_of_dictionaries.append(row_list)
```
